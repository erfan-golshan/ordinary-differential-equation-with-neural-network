```python
import torch
from torch import nn

import matplotlib.pyplot as plt

import numpy as np
```

```python

model = nn.Sequential(
    nn.Linear(1, 4),
    nn.Sigmoid(),
    nn.Linear(4, 1),
)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.SGD(model.parameters(),lr=0.05,momentum=0.9)
```

```python
loss_list=[]

# dy/dx = -1/(x-5)**2 + 2x, y(0) = 0.4

x = np.linspace(0,1,100)

for i in range(len(x)-1):
    y0 = model(torch.tensor([0.0]))
    xa = float(x[i])
    xb = float(x[i+1])
    ya = model(torch.tensor([xa]))
    yb = model(torch.tensor([xb]))

    dy = (yb-ya)/(xb-xa)

    loss1 = criterion(dy, torch.tensor([-1/(xb-5)**2 + 2*xb]))
    loss2 = criterion(y0, torch.tensor([0.4]))
    loss = loss1 + loss2

    ####################
    loss_list.append(loss.item())
    ####################
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

plt.plot(loss_list)
```

```python
x = np.linspace(0,1,100)
y=[]
for i in range(len(x)):
    y.append(model(torch.tensor([float(x[i])]))[0].item())


plt.rcParams["figure.figsize"] = [5, 4]
plt.rcParams["figure.autolayout"] = True
line1, = plt.plot(x , 1/(x-5) + x**2 + 0.6 , label="f")
line2, = plt.plot(x , y , label="nn")

leg = plt.legend(loc='upper center')
plt.show()
```
